```python
import json
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from plotly.io import read_json
```

```python
import pickle
with open('27099553/df.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python

x = data.index.to_list()
z = data.values

adict = {}
index = 0
for i in range(0, len(x)):
    temp_z = z[i]
    flag = 0
    for k in range(0, len(temp_z)):
        if temp_z[k] > 2:
            flag = 1
            break
    if flag == 0:
        adict[f'x{index+1}'] = z[i].tolist()
        index += 1

df = pd.DataFrame(adict)

fig = go.Figure(data=[go.Surface(z=df.values, x=x)])

fig.update_layout(title='3D Surface', autosize=False,
                  width=900, height=500,
                  scene=dict( bgcolor="seagreen", aspectratio=dict(x=4, y=1, z=0.7), xaxis= {"nticks": 40}))
fig.update_xaxes(nticks=1)
fig.show()
```
